```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
returns = np.random.normal(0.001,0.2,250)    #centre of distrubution,standart deviasion,size
```

```python
returns[:20]
```

```python
initial_price = 100
```

```python
price = initial_price * np.exp(returns.cumsum())
```

```python
plt.plot(price)
plt.grid();
```
